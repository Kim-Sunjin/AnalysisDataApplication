```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Workspace/분석데이터활용/pandasData분석/data/zoom채팅.csv', header=None)
```

```python
values = df.values
values
```

```python
re_list = []

for i in values:
    if ":" in i[0]:
        line = []
        line.append(i[0])
    else:
        line.append(i[0])
    re_list.append(line)
```

```python
re_list
```

```python
name = []
for i in re_list:
    tmp = i[0].split(' ')[0]

    tmp1 = tmp.split('_')
    if len(tmp1)>=2:
        name.append(tmp1[1])
```

```python
from collections import Counter
Counter(name).most_common(5)
```
