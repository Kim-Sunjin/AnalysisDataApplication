```python
X_train=add_constant(X_train)
X_test=add_constant(X_test)
ols_model=OLS(y_train,X_train)
results=ols_model.fit()
results.summary()
```

```python
print(results.predict(X_test)[0:5]) # 실수형
print('------------')
print(y_test[0:5]) # 정수형
```

```python
# 반올림 함수 (예측값과 결과값의 형식을 같게 하기 위함)
def rounder(array):
    rounded_array=[]
    for i in array:
        rounded_array.append(int(i))
    return np.array(rounded_array)
```

```python
ols_pred=results.predict(X_test)
rounded_pred_ols=rounder(ols_pred)
# 편차값, rmse는 낮을수록 좋음
print(f'The RMSE of sklearn model: {np.sqrt(mean_squared_error(rounded_pred_ols,y_test))}')
```

### 트레이닝
- 테스트 데이터 분리
- 표준화
- 회귀분석 rmse값 출력
- 로지스틱회귀 rmse값 출력

```python
y=df['quality']
X=df.drop(['quality','total sulfur dioxide'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
print(X_train.shape, X_test.shape,y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import Lasso
alpha = 10
model_las = Lasso(alpha=alpha) # Lasso 이용
model_las.fit(X_train, y_train)
pred_las = model_las.predict(X_test)
rounded_pred_las = rounder(pred_las)
print(model_las.score(X_test,y_test))
print(f'The RMSE of sklearn model: {np.sqrt(mean_squared_error(rounded_pred_las,y_test))}')
```

```python
# 로지스틱회귀(scaling을 한 것)
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)
model_log = LogisticRegression(penalty='none')
model_log.fit(train_scaled, y_train)
pred_log = model_log.predict(test_scaled)
rounded_pred_log = rounder(pred_log)
print(f'The RMSE of sklearn model: {np.sqrt(mean_squared_error(rounded_pred_log,y_test))}')
```
